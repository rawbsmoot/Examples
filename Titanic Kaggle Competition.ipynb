```python
# importing packages we will be using.

import numpy as np
import pandas as pd 
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, roc_curve, auc, f1_score
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from patsy import dmatrices
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline, make_pipeline, make_union
from sklearn import metrics
from sklearn.preprocessing import Imputer, StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from IPython.display import Image
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import roc_auc_score, classification_report
from sqlalchemy import create_engine
import patsy
from sklearn import neighbors, metrics
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
```

```python
# defining some funcions to help us plot.

def auc_plotting_function(rate1, rate2, rate1_name, rate2_name, curve_name):
    AUC = auc(rate1, rate2)
    # Plot of a ROC curve for class 1 (survived)
    plt.figure(figsize=[11,9])
    plt.plot(rate1, rate2, label=curve_name + ' (area = %0.2f)' % AUC, linewidth=4)
    plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel(rate1_name, fontsize=18)
    plt.ylabel(rate2_name, fontsize=18)
    plt.title(curve_name + ' Survival Prediction', fontsize=18)
    plt.legend(loc="lower right")
    plt.show()
    
def plot_roc(y_test, y_score):
    fpr, tpr, _ = roc_curve(y_test, y_score)
    auc_plotting_function(fpr, tpr, 'False Positive Rate', 'True Positive Rate', 'ROC')
    
def plot_confusion_matrix(cm, title='Confusion Matrix', cmap=plt.cm.Greens):
    plt.imshow(cm, interpolation='nearest', cmap = cmap)
    plt.title(title)
    plt.colorbar()
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

# <center>Logistic Regression Model</center>

```python
X = df.ix[:,[5, 8, 10, 12, 13, 14, 15, 16]]
y = df['Survived'].ravel()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, stratify = y, random_state = 30)
```

```python
# In this step we will instantiate our model, fit our model and predict survival with our test set.  We will also look
# at our coefficients.  It looks like most of them are better at indicating if you will die than if you will survive.

lr = LogisticRegression(random_state = 30, penalty ='l1')
logreg = lr.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
model_coefs = zip(X.columns, np.transpose(lr.coef_))
model_coefs
```

```python
# Let's see our scores.  

print 'Accuracy: ', accuracy_score(y_test, y_pred)
print 'Precision: ', precision_score(y_test, y_pred)
print 'Recall: ', recall_score(y_test, y_pred)
print 'F1-Score: ', f1_score(y_test, y_pred)
```

```python
# Here is our classification report.  This is similar to the scores we looked at previously but more in-depth.

target_names = ['Died', 'Survived']
print classification_report(y_test, y_pred, target_names = target_names)
```

```python
# Now let's try a cross evaluation using 5 folds to see if our scores are reliable.  It's looking pretty good.  The 
# mean score is quite close to our original score of 0.8.

cv_score = cross_val_score(lr, X, y, cv = 5)
print 'Cross Validation Score Mean: ', cv_score.mean()
print 'Cross Validation Score StandDv: ', cv_score.std()
print cv_score
```

```python
# Below we have created a confusion matrix showing our outcomes.

cm = np.array(confusion_matrix(y_test, y_pred))
df_cm = pd.DataFrame(cm, index=['Died', 'Survived'],
                         columns=['Predicted_Died','Predicted_Survived'])
df_cm
```

```python
# Here is the same thing but graphically representd.
plt.figure()
plt.grid(b=None)
plot_confusion_matrix(df_cm)
```

```python
# We will also look at an ROC curve.  This is our True Positive Rate by our False Positive Rate.  We can see that as we
# approach %80 True Positive Rate we begin to flatten out.  With this model we will only be able to predict about %80 of
# passenger outcomes (death, survival)

y_score = logreg.decision_function(X_test)
plot_roc(y_test, y_score)
plt.show()
```

```python
# Our confusion matrix looks much better!

y_pp = pd.DataFrame(logreg.predict_proba(X_test), columns=['Died','Survived'])
y_pp['Threshold'] = [1 if x >= .65 else 0 for x in y_pp.Survived.values]
cm_alt = np.array(confusion_matrix(y_test, y_pp.Threshold))
confusion_alt = pd.DataFrame(cm_alt, index=['Died', 'Survived'],
                         columns=['Predicted_Died','Predicted_Survived'])

confusion_alt
```

```python
# Let's check our scores.  Sheesh!  We've improved our model by more than 3% by changing the decision threshold!!

print 'Accuracy: ', accuracy_score(y_test, y_pp.Threshold)
print 'Precision: ', precision_score(y_test, y_pp.Threshold)
print 'Recall: ', recall_score(y_test, y_pp.Threshold)
print 'F1-Score: ', f1_score(y_test, y_pp.Threshold)
```

# <center>Using GridSearchCV with Logistic Regression</center>

```python
# We are going to try several parameters to find the best model.  We initially tried it with L1 regularization but no
# models were better than the models that used L2 regularization.  Below we set our parameters and run our grid search.

logreg_parameters = {
    'penalty':['l2'],
    'C':np.array([10,1,0.1,0.01,0.001,0.0001]),
    'solver':['liblinear', 'newton-cg', 'sag', 'lbfgs']}
log = LogisticRegression()
log_grid = GridSearchCV(log, param_grid=logreg_parameters, cv=5)
log_grid.fit(X, y)
logreg_pred = log_grid.predict(X_test)
```

```python
# Let's see what our grid search suggests for us to use in order to create a superior model.

print log_grid.best_score_
print log_grid.best_params_
print log_grid.best_estimator_
```
